#### Combine these dataframe into one 

```python
#df_uk = pd.merge(df_uk_totall_case,df_uk_totall_death,df_uk_totall_test)
df_uk = pd.concat([df_uk_totall_case,df_uk_totall_death,df_uk_totall_test], axis=1, sort=False)
df_uk
```

```python
Headers = ['Total_Case','Total_Death','Total_Test']
Indexs = ['England','Northern Ireland','Scotland','Wales']
df_uk.columns = Headers
df_uk.index = Indexs
df_uk
```

```python
df_uk = pd.concat([df_uk,df_location], axis=1)
df_uk
```

#### (2).Draw the map and mark it.

###### Check the type

```python
df_uk.dtypes
```

###### The latitude and longitude should not be object. So use '.astype()' method change the type to float. As the same the number should be shown which should be object/str.

```python
df_uk['Latitude'] = df_uk['Latitude'].astype('float')
df_uk['Longitude'] = df_uk['Longitude'].astype('float')
df_uk['Total_Case'] = df_uk['Total_Case'].astype('str')
df_uk['Total_Death'] = df_uk['Total_Death'].astype('str')
df_uk['Total_Test'] = df_uk['Total_Test'].astype('str')
df_uk.dtypes
```

#### Drwa the case map

```python
# let's start again with a clean copy of the map of San Francisco
map_uk_case = folium.Map(location = [uk_latitude, uk_longitude], zoom_start = 5.5)

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(map_uk_case)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(df_uk.Latitude, df_uk.Longitude, df_uk.Total_Case):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# display map
map_uk_case
```

#### Draw the death map

```python
# let's start again with a clean copy of the map of San Francisco
map_uk_death = folium.Map(location = [uk_latitude, uk_longitude], zoom_start = 5.5)

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(map_uk_death)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(df_uk.Latitude, df_uk.Longitude, df_uk.Total_Death):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# display map
map_uk_death
```
